```python
def scrape_page(page_number):
    url = f'https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst&start={page_number}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    containers = soup.findAll("div", {"class": "css-1gatmva"})

    jobs = []
    
    for container in containers:
        # job title
        jtitle = container.findAll("h2", {"class": "css-m604qf"})
        job_title = jtitle[0].text.strip() if jtitle else 'N/A'

        # company name
        cname = container.findAll("a", {"class": "css-17s97q8"})
        company_name = cname[0].text.strip() if cname else 'N/A'

        # job type
        jtype = container.findAll("span", {"class": "css-1ve4b75 eoyjyou0"})
        job_type = jtype[0].text.strip() if jtype else 'N/A'

        #site
        jsite= container.findAll("span", {"class": "css-o1vzmt eoyjyou0"})
        jsite = jsite[0].text.strip() if jsite else 'N/A'

        

        #store
        jobs.append([job_title, company_name, job_type, jsite])
    
    return jobs
```

```python
import csv
import requests
from bs4 import BeautifulSoup
```

```python
with open('WEbScraping.csv', 'w', newline='', encoding='utf-8') as csvfile:
    # CSV
    csvwriter = csv.writer(csvfile)

    # header row
    csvwriter.writerow(['Job Title', 'Company Name', 'Job Type','Job Site'])

    # scrap 20 page
    for page_number in range(0, 20):
        jobs = scrape_page(page_number)
        for job in jobs:
            csvwriter.writerow(job)
        print(f'Scraped page {page_number + 1}')
```
